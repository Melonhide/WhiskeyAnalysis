### 754 is not in the whiskey list

```python
whiskey_avg_rating = df_rating.groupby('whiskey')['Liking - DEPENDANT VARIABLE'].mean()
whiskey_avg_rating.sort_values()
```

```python
highest = len(whiskey_avg_rating)-1
lowest = 0
twenty_five = highest//4
middle = highest//2
seventy_five = highest//4*3

sample_whiskey = [lowest, twenty_five, middle, seventy_five, highest]
sample_whiskey
```

```python
whiskey_avg_rating.sort_values().iloc[sample_whiskey]
```

### Let's pick 948, 724, 494, 663, 135

```python
sample_whiskey_id = whiskey_avg_rating.sort_values().iloc[sample_whiskey].index
sample_whiskey_ratings = df_rating[df_rating['whiskey'].isin(sample_whiskey_id)]
```

```python
sample_whiskey_ratings['whiskey'].value_counts()
```

```python
sample_whiskey_ratings
```

```python
sns.countplot(hue='whiskey', x='Liking - DEPENDANT VARIABLE', data=sample_whiskey_ratings)
plt.xlabel('Whiskey')
plt.ylabel('Count')
plt.title('Rating Counts of Whiskies')
plt.show()
```

```python
sns.set(style="whitegrid")

for whiskey in sample_whiskey_ratings['whiskey'].unique():
    plt.figure()
    sns.countplot(x='Liking - DEPENDANT VARIABLE', data=sample_whiskey_ratings[sample_whiskey_ratings['whiskey'] == whiskey])
    plt.title(f'Rating Counts for Whiskey {whiskey}')
    plt.xlabel('Rating')
    plt.ylabel('Count')
    plt.show()
```

```python
df_whiskey.columns
```

```python
df_merged = pd.merge(df_rating, df_whiskey, left_on='whiskey', right_on='Whiskey').drop('Whiskey', axis = 1)
df_merged
```

```python
correlation_matrix = df_merged.drop(['UID', 'whiskey'], axis = 1).corr()
# 可视化相关性矩阵
plt.figure(figsize=(10, 8))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm')
plt.title('Correlation between Ratings and Flavor Profiles')
plt.show()
```
